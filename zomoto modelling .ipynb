```python
from sklearn.model_selection import learning_curve
import numpy as np

train_sizes, train_scores, test_scores = learning_curve(
    SVC(kernel='rbf', C=1, gamma='scale', random_state=42),
    X, y,
    cv=5,
    scoring='accuracy',
    n_jobs=-1
)

train_mean = np.mean(train_scores, axis=1)
test_mean = np.mean(test_scores, axis=1)

plt.figure(figsize=(8,5))
plt.plot(train_sizes, train_mean, 'o-', label='Training Accuracy')
plt.plot(train_sizes, test_mean, 'o-', label='Validation Accuracy')
plt.title("Bias–Variance Tradeoff (SVM)")
plt.xlabel("Training Set Size")
plt.ylabel("Accuracy")
plt.legend()
plt.grid(True)
plt.show()
```

```python
from sklearn.svm import SVR
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline

# ✅ Create and fit the SVM Regression pipeline
svm_reg = make_pipeline(StandardScaler(), SVR(kernel='rbf', C=10, gamma='scale'))
svm_reg.fit(X_train, y_train)

# ✅ Evaluate model performance
train_score = svm_reg.score(X_train, y_train)
test_score = svm_reg.score(X_test, y_test)

print(f"Train R²: {train_score:.3f}")
print(f"Test R²:  {test_score:.3f}")

# ✅ Bias–Variance analysis
if abs(train_score - test_score) < 0.05:
    print("✅ Model is balanced (good bias–variance tradeoff).")
elif train_score > test_score:
    print("⚠️ High variance (overfitting). Try reducing model complexity or regularization (lower C).")
else:
    print("⚠️ High bias (underfitting). Try higher C or nonlinear kernel (RBF).")
```

```python
#Lower the Regularization Parameter C
for C in [0.1, 1, 5]:
    svm_reg = make_pipeline(StandardScaler(), SVR(kernel='rbf', C=C, gamma='scale'))
    svm_reg.fit(X_train, y_train)
    print(f"C={C}, Train R²: {svm_reg.score(X_train, y_train):.3f}, Test R²: {svm_reg.score(X_test, y_test):.3f}")
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.svm import SVR

param_grid = {
    'svr__C': [0.1, 1, 10],
    'svr__gamma': ['scale', 0.1, 0.01],
    'svr__epsilon': [0.05, 0.1, 0.2]
}

svm_reg = make_pipeline(StandardScaler(), SVR(kernel='rbf'))
grid = GridSearchCV(svm_reg, param_grid, cv=5, scoring='r2', n_jobs=-1, verbose=2)
grid.fit(X_train, y_train)

print("Best Parameters:", grid.best_params_)
print("Best Cross-Validation R²:", grid.best_score_)

# Evaluate on test set
best_svm = grid.best_estimator_
print("Test R²:", best_svm.score(X_test, y_test))
```
